```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
dummy_df = pd.get_dummies(df["Car Model"],dtype=int)
dummy_df
```

```python
mergerdf = pd.concat([df, dummy_df],axis= 'columns')
mergerdf
```

```python
final_df = mergerdf.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final_df
```

```python
%matplotlib inline
plt.ylabel('Sell Price($)')
plt.xlabel('Mileage')
plt.scatter(final_df["Mileage"],df['Sell Price($)'],color='blue')
```

```python
%matplotlib inline
plt.ylabel('Sell Price($)')
plt.xlabel('Age(yrs)')
plt.scatter(final_df["Age(yrs)"],df['Sell Price($)'],color='blue')
```

```python
inputdf = final_df.drop(["Sell Price($)"],axis='columns')
inputdf
```

```python
from sklearn.linear_model import LinearRegression
```

```python
req = LinearRegression()
```

```python
req.fit(inputdf,final_df['Sell Price($)'])
```

```python
req.predict([[45000,4,0,0]])
```

```python
req.predict([[86000,7,0,1]])
```

```python
req.score(inputdf,final_df['Sell Price($)'])
```

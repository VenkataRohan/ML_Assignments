```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
plt.matshow(X_train[2])
```

```python
X_train1 = X_train/255
X_train1[0]
y_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(25, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


model.fit(X_train1, y_train, epochs=20)
```

```python
cnn = keras.Sequential([
    
    #cnn
    keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28,1)),
    keras.layers.MaxPooling2D((2, 2)),
    
    
    #ann
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


cnn.fit(X_train1, y_train, epochs=20)
```

```python
X_test =X_test/255
X_test[0]
```

```python
model.evaluate(X_test,y_test)
```

```python
cnn.evaluate(X_test,y_test)
```

```python
y_test_predict = model.predict(X_test)
np.argmax(y_test_predict[0])
```

```python
y_test[0]
```

```python
y_test_predicted_label = [np.argmax(i) for i in y_test_predict]
y_test_predicted_label
```

```python
cf = tf.math.confusion_matrix(labels=y_test,predictions=y_test_predicted_label)
cf
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cf, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

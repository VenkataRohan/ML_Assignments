```python
from pypdf import PdfReader

reader = PdfReader("test.pdf")

# Print the number of pages in the PDF
print(f"There are {len(reader.pages)} Pages")

# Get the first page (index 0) 
page = reader.pages[0]
# Use extract_text() to get the text of the page
print(page.extract_text())

# Go through every page and get the text
for i in range(len(reader.pages)):
  page = reader.pages[i]
  print(page.extract_text())
```

```python
for img in page.images:
    with open(img.name, "wb") as fp:
        fp.write(img.data)
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)

    for page_number in range(doc.page_count):
        page = doc[page_number]
        text = page.get_text()
        print(f"Page {page_number + 1}:\n{text}\n")

    doc.close()

# Example usage
pdf_file_path = "test.pdf"
extract_text_from_pdf(pdf_file_path)
```

```python
import tabula

def extract_tables_from_pdf(pdf_path):
    tables = tabula.read_pdf(pdf_path, pages='all')
    for i, table in enumerate(tables, start=1):
        print(f"Table {i}:\n{table}\n")

# Example usage
pdf_file_path = "test.pdf"
extract_tables_from_pdf(pdf_file_path)
```

```python
import camelot

def extract_tables_from_pdf(pdf_path):
    tables = camelot.read_pdf(pdf_path, flavor='stream', pages='all')
    for i, table in enumerate(tables, start=1):
        print(f"Table {i}:\n{table.df}\n")

# Example usage
pdf_file_path = "test.pdf"
extract_tables_from_pdf(pdf_file_path)
```
